# 导入两个包numpy是数学工具，matplotlib.pyplot是画图
## np.random.rand()生成一个随机数
## plt.scatter(x,y)可以画图

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(100)
z = np.random.rand(100)
print(x)
y = x
z = x**2
plt.scatter(x,y)
plt.scatter(x,z)
f = 2*2.7**(-((x-1)**2/2*3**2))
plt.scatter(x,f)
```
